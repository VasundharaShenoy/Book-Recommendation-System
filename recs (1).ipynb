```python
import pandas as pd
data = pd.read_csv('./Data/goodreads.csv')
data.head()
```

```python
print(data.shape)
df = data[['genre', 'title', 'rate', 'book_link']].copy()
df.head()
```

```python
df['genre'] = df['genre'].astype(str)
```

```python
genres_all_books = [df.loc[i]['genre'].split(',') for i in df.index]
# Find the list of genres of all movies in alphabetical order
genres = sorted(list(set([item for sublist in genres_all_books for item in sublist])))
genres
```

```python
len(genres)
```

```python

full_data = []
book_titles = []
```

```python
print(df)
```

```python
df.to_csv('test.csv')
```

```python
for i in df.index:
    # Append book title and the index of the book
    book_titles.append((df.loc[i]['title'].strip(), i, df.loc[i]['book_link'].strip()))
    # Add list of genres of the book (1/0) to book data
    book_data = [1 if genre in df.loc[i]['genre'].split(',') else 0 for genre in genres]
    # Add rating of the book to the book data
    book_data.append(df.loc[i]['rate'])
    # Add record of book to main data list
    full_data.append(book_data)
```

```python
import json

data_dump = './Data/book_data.json'
titles_dump = './Data/book_title.json'
with open(data_dump, 'w+', encoding='utf-8') as f:
    json.dump(full_data, f)
with open(titles_dump, 'w+', encoding='utf-8') as f:
    json.dump(book_titles, f)
```
